## Importando dados

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

## Relembrando aula anterior

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df['comprimento_pétala'], bins=20, density=True)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

## Adiconando média e desvio padrão
Também é removido a densidade de probabilidade do histograma

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

# Media e desvio padrão
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

# String para formatar
eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma), xy=(5.4,30), fontsize=15)
```

## Adicionando linhas de media e mediana

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma), xy=(5.4,30), fontsize=15)

# Adiciona linha de média e texto indicando
eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-1.2, 28), fontsize=15)

# Adiciona linha de mediana e texto indicando
eixo.axvline(df['comprimento_pétala'].median(), color='g', linestyle='-.')
eixo.annotate('Mediana', xy=(df['comprimento_pétala'].median()+.3, 25), fontsize=15, color='green')
```

## Alterando anotações para representar simbolos gregos

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

# Similar à LaTeX
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), xy=(5.4,30), fontsize=15)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-1.2, 28), fontsize=15)

eixo.axvline(df['comprimento_pétala'].median(), color='g', linestyle='-.')
eixo.annotate('Mediana', xy=(df['comprimento_pétala'].median()+.3, 25), fontsize=15, color='green')
```

### (Extra) Histograma com boxplot juntos

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, .5])
eixo2 = fig.add_axes([0, .5, 1, .5])
eixo2.boxplot(df['comprimento_pétala'],vert=False)

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

# Similar à LaTeX
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), xy=(5.4,15), fontsize=15)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-1.2, 28), fontsize=15)

eixo.axvline(df['comprimento_pétala'].median(), color='g', linestyle='-.')
eixo.annotate('Mediana', xy=(df['comprimento_pétala'].median()+.3, 25), fontsize=15, color='green')
```

## Agrupando por espécie e salvando imagem

### Iris Versicolor

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

# Dataframe Iris versicolor
df_iv = df[df['espécie']=='Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins=20)
eixo.set_title('Iris-versicolor', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), xy=(4.6,6), fontsize=15)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-0.3, 5.5), fontsize=15)

eixo.axvline(df_iv['comprimento_pétala'].median(), color='g', linestyle='-.')
eixo.annotate('Mediana', xy=(df_iv['comprimento_pétala'].median()-0.5, 6.5), fontsize=15, color='green')

## Salvando imagem
fig.savefig('histograma_iv.png', bbox_inches='tight')
```

### Iris Setosa 

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

# Dataframe Iris Setosa
df_is = df[df['espécie']=='Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins=20)
eixo.set_title('Iris-setosa', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), xy=(1,12), fontsize=15)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-0.15, 13), fontsize=15)

eixo.axvline(df_is['comprimento_pétala'].median(), color='r', linestyle='-.')
eixo.annotate('Mediana', xy=(df_is['comprimento_pétala'].median()+0.05, 13), fontsize=15, color='r')

## Salvando imagem
fig.savefig('histograma_is.png', bbox_inches='tight')
```

### Iris Virginca

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

# Dataframe Iris Virginica
df_iva = df[df['espécie']=='Iris-virginica']

mu, sigma = df_iva['comprimento_pétala'].mean(), df_iva['comprimento_pétala'].std()

eixo.hist(df_iva['comprimento_pétala'], bins=20)
eixo.set_title('Iris-virginica', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), xy=(4.4,8), fontsize=15)

eixo.axvline(mu, color='k', linestyle='--')
eixo.annotate('Média', xy=(mu-.35, 8), fontsize=15)

eixo.axvline(df_iva['comprimento_pétala'].median(), color='r', linestyle='-.')
eixo.annotate('Mediana', xy=(df_iva['comprimento_pétala'].median()+0.05, 8), fontsize=15, color='r')

## Salvando imagem
fig.savefig('histograma_iva.png', bbox_inches='tight')
```

# Combinando as 3 imagens em uma só

### Importando Image da biblioteca Python Image Library

```python
from PIL import Image
```

```python
# Pega altura e largura das imagens
largura, altura = Image.open('histograma_is.png').size

# Cria nova imagem e intervalo para servir de contador
imagem_combinada = Image.new('RGB',size=(3*largura-10, altura))
intervalo = 0

# Abre as imagens e itera por cada uma, colando elas lado a lado
for index, imagem in enumerate(map(Image.open, ['histograma_is.png', 'histograma_iv.png', 'histograma_iva.png'])):
    imagem_combinada.paste(imagem, (intervalo, 0))

    # Esse if serve apenas para evitar uma borda preta que aparece entra as imagens porque elas não exatamente do mesmo tamanho
    if index < 1:
        intervalo += largura
    else:
        intervalo += largura - 10


imagem_combinada.save('combinada.png')
imagem_combinada
```
