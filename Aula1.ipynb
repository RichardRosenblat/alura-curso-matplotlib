## Fazendo primeiras visualizações

### Importando e pré processando dados

```python
import pandas as pd
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Importante passar a data de string para datetime
df['data'] = pd.to_datetime(df['data'])
```

```python
import matplotlib.pyplot as plt
```

### Grafico simples

```python
plt.plot(df['data'], df['temperatura'])
```

### Ajustando tamanho

```python
plt.figure(figsize=(15, 8))
plt.plot(df['data'], df['temperatura'])
```

### Adiconando titulo

```python
plt.figure(figsize=(15, 8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura X Tempo')
```

### Criando nova figura e desenhando eixos para mais controle do grafico

```python
fig = plt.figure(figsize=(12, 5))
eixo = fig.add_axes([0, 0, 1, 1])
```

### Adicionando dados e titulo

```python
fig = plt.figure(figsize=(12, 5))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
```

### Adicionando labels nos eixos

```python
fig = plt.figure(figsize=(12, 5))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
eixo.set_ylabel('Temperatura')
eixo.set_xlabel('Data')
```

### Mudando tamanho de textos

```python
fig = plt.figure(figsize=(12, 5))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', fontsize=20,)
eixo.set_ylabel('Temperatura',fontsize=15)
eixo.set_xlabel('Data',fontsize=15)
eixo.legend(['Temperatura'], loc='lower right', fontsize=12)
```

### Adicionando legenda

```python
fig = plt.figure(figsize=(12, 5))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura',fontsize=15)
eixo.set_xlabel('Data',fontsize=15)
eixo.legend(['Temperatura'], loc='lower right', fontsize=12)
```

### Alterando cores

```python
fig = plt.figure(figsize=(12, 5))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'],color='green')

eixo.set_title('Temperatura no momento', fontsize=20)
eixo.set_ylabel('Temperatura',fontsize=15)
eixo.set_xlabel('Data',fontsize=15)
eixo.legend(['Temperatura'], loc='lower right', fontsize=12)
```
