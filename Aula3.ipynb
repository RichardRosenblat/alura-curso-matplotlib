### Importando e pré processando dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
df['data'] = pd.to_datetime(df['data'])
df.head()
```

### Relembrando aula anterior

```python
fig = plt.figure(figsize=(12, 5))

eixo = fig.add_axes([0, 0, 1, 1])
eixo_2 = fig.add_axes([.6, .66, .4, .3])

eixo.grid(True)
eixo.plot(df['data'],df['temperatura'], color='g')
eixo.set_xlim(datetime.datetime(2014, 5, 1), datetime.datetime(2014, 6, 1))
eixo.set_ylim(270, 320)
eixo.set_title('Temperatura em maio/2014', fontsize=20, pad=15)
eixo.legend(['Temperatura'], loc='lower right', fontsize=12)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.set_xlabel('Data', fontsize=15)

azul_esquerda = df['data'] < datetime.datetime(2014, 5, 1)
azul_direita = df['data'] > datetime.datetime(2014, 6, 1)

eixo_2.grid(True)
eixo_2.plot(df['data'], df['temperatura'], color='g')
eixo_2.plot(df[azul_esquerda]['data'], df[azul_esquerda]['temperatura'], color='b')
eixo_2.plot(df[azul_direita]['data'], df[azul_direita]['temperatura'], color='b')
eixo_2.set_xlim(datetime.datetime(2014, 1, 1), datetime.datetime(2015, 1, 1))
eixo_2.set_title('Temperatura em 2014', fontsize=10, pad=15)
eixo_2.legend(['Temperatura'], loc='best', fontsize=8)
eixo_2.set_ylabel('Temperatura', fontsize=8)
eixo_2.set_xlabel('Data', fontsize=8)
```

### Recriando Gráfico de temperatura no momento e adicionando limitadores

```python
# Recriando grafico
fig = plt.figure(figsize=(12,5))

eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_title('Temperatura no momento', fontsize=20, pad=15)
eixo.set_xlabel('Data', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.legend(['Temperatura'], loc='lower right',  fontsize=12)
eixo.grid(True)

# Criando limitadores

eixo.axhline(max(df['temperatura']), color='r', ls='--')
eixo.axhline(min(df['temperatura']), color='c', ls='--')
```

### Adicionando notação em limitadores de máximo

```python
# Recriando grafico
fig = plt.figure(figsize=(12,5))

eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_title('Temperatura no momento', fontsize=20, pad=15)
eixo.set_xlabel('Data', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.legend(['Temperatura'], loc='lower right',  fontsize=12)
eixo.grid(True)

# Criando limitadores
eixo.axhline(max(df['temperatura']), color='r', ls='--')
eixo.axhline(min(df['temperatura']), color='c', ls='--')

# Data de quando ocorreu a maior temperatura
max_label_x = df['data'][df['temperatura'].idxmax()]
# Maior temperatura
max_label_y = max(df['temperatura'])

# Escreve no grafico
eixo.annotate('Máximo', xy=(max_label_x, max_label_y), fontsize=15)
```

### Melhorando notações

```python
# Recriando grafico
fig = plt.figure(figsize=(12,5))

eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_title('Temperatura no momento', fontsize=20, pad=15)
eixo.set_xlabel('Data', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.legend(['Temperatura'], loc='lower right',  fontsize=12)
eixo.grid(True)

# Criando limitadores
eixo.axhline(max(df['temperatura']), color='r', ls='--')
eixo.axhline(min(df['temperatura']), color='c', ls='--')


# Data de quando ocorreu a maior temperatura
max_temp_x = df['data'][df['temperatura'].idxmax()]
# Maior temperatura
max_temp_y = max(df['temperatura'])

# Pega o dia da temperatura mais alta e subtrai 7000 horas do resultado
max_temp_text_x = df['data'][df['temperatura'].idxmax() - 7000]
max_temp_text_y = max(df['temperatura'] - 5)

# Escreve no grafico e cria a seta
eixo.annotate(
    'Máximo', 
    xy=(max_temp_x, max_temp_y), 
    fontsize=15,
    xytext=(max_temp_text_x, max_temp_text_y), 
    arrowprops={'facecolor':'r'}
)
```

### Adicionando notação em limitadores de mínimo

```python
# Recriando grafico
fig = plt.figure(figsize=(12,5))

eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color='g')
eixo.set_title('Temperatura no momento', fontsize=20, pad=15)
eixo.set_xlabel('Data', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.legend(['Temperatura'], loc='lower right',  fontsize=12)
eixo.grid(True)

# Criando limitadores
eixo.axhline(max(df['temperatura']), color='r', ls='--')
eixo.axhline(min(df['temperatura']), color='c', ls='--')


# Data de quando ocorreu a maior temperatura
max_temp_x = df['data'][df['temperatura'].idxmax()]
# Maior temperatura
max_temp_y = max(df['temperatura'])

# Pega o dia da temperatura mais alta e subtrai 7000 horas do resultado
max_temp_text_x = df['data'][df['temperatura'].idxmax() - 7000]
max_temp_text_y = max(df['temperatura'] - 5)

# Escreve no grafico e cria a seta
eixo.annotate(
    'Máximo', 
    xy=(max_temp_x, max_temp_y), 
    fontsize=15,
    xytext=(max_temp_text_x, max_temp_text_y), 
    arrowprops={'facecolor':'r'}
)


# Mesmo processo anterior, agora para a menor temperatura
min_temp_x = df['data'][df['temperatura'].idxmin()]
min_temp_y = min(df['temperatura'])
min_temp_text_x = df['data'][df['temperatura'].idxmin() - 6000]
min_temp_text_y = min(df['temperatura'] + 5)
eixo.annotate(
    'Mínimo', 
    xy=(min_temp_x, min_temp_y), 
    fontsize=15,
    xytext=(min_temp_text_x, min_temp_text_y), 
    arrowprops={'facecolor':'c'}
)
```

### Agrupando por dia da semana

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()

dias_das_semana = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[dias_das_semana]

temperatura_por_dia_da_semana
```

### Adicionando dados agrupados por dia da semana em um grafico de barras

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.bar(dias_das_semana, temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
```

#### Forma alternativa

```python
### Forma alternativa de fazer esse gráfico é:
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

indice = range(len(temperatura_por_dia_da_semana))

eixo.bar(indice, temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)

# Só pra demonstrar que o eixo x funciona com números também
eixo.set_xticks(indice)
eixo.set_xticklabels(dias_das_semana)
```

### Adicionando cores às barras

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

cores = ['green', 'yellow', 'purple',  'cyan', 'blue','orange', 'red']

# Edge color é a cor da borda das cores
eixo.bar(dias_das_semana, temperatura_por_dia_da_semana, color=cores, edgecolor='k')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
```

### Mexendo com limites e escalas

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

cores = ['green', 'yellow', 'purple',  'cyan', 'blue','orange', 'red']


eixo.bar(dias_das_semana, temperatura_por_dia_da_semana, color=cores, edgecolor='k')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)

# Definindo um periodo bem pequeno ao gráfico
eixo.set_ylim(284.8,285.1)
```

### Grafico de pizza

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

### Adicionando porcentagens

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index, autopct='%.1f%%')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

### explodindo grafico de pizza

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

explodir = [.1, 0, 0, 0, 0, .1, .1]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index, autopct='%.1f%%', explode=explodir)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

### Adicionando sombra ao grafico de pizza

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

explodir = [.1, 0, 0, 0, 0, .1, .1]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index, autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```
