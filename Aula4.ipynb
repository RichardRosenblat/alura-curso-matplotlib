## Importando dados

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

## Scattering os dados

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Grafico de dispersão', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
```

## Aumentando tamanho dos ticks

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Grafico de dispersão', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)

# Configurando ticks
eixo.tick_params(labelsize=12)
```

## Alterando cor por espécie

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0, 0, 1, 1])

# Definimos cores
cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}

# Iteramos por cada espécie no dataframe
for especie in df['espécie'].unique():
    # Selecionamos flores da especie que estamos iterando
    temp = df[df['espécie'] == especie]
    # Colocamos os dados selecionados e especificamos as cores selecionadas
    eixo.scatter(temp['comprimento_sépala'], temp['largura_sépala'], color=cores[especie])

# Adicionamos uma legenda especificando as cores
eixo.legend(cores, fontsize=13)

eixo.set_title('Grafico de dispersão', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)

eixo.tick_params(labelsize=12)
```

## Alterando marcadores por espécie

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}

#Definimos marcadores
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

for especie in df['espécie'].unique():
    temp = df[df['espécie'] == especie]
    # Setamos os marcadores por espécie
    eixo.scatter(temp['comprimento_sépala'], temp['largura_sépala'], color=cores[especie], marker=marcadores[especie])

eixo.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], fontsize=13)

eixo.set_title('Grafico de dispersão', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)

eixo.tick_params(labelsize=12)
```

## Alterando tamanho de marcadores 

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica':'g'}
marcadores = {'Iris-setosa':'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

for especie in df['espécie'].unique():
    temp = df[df['espécie'] == especie]
    eixo.scatter(
        temp['comprimento_sépala'], 
        temp['largura_sépala'], 
        color=cores[especie], 
        marker=marcadores[especie],
        s=100
    )

eixo.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], fontsize=13)

eixo.set_title('Grafico de dispersão', fontsize=20, pad=10)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)

eixo.tick_params(labelsize=12)
```

## Usando boxplot para visualisar a distribuição da largura das pétalas

Mais informações sobre boxplots [aqui](https://www.robrac.org.br/seer/index.php/ROBRAC/article/download/1132/897/)

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.boxplot(df['largura_pétala'])
eixo.set_title('Distribuição da largura das Pétalas', fontsize=15, pad=10)

eixo.set_xticklabels(['Largura das petalas'])
```

## Alterando boxplot para mostrar todas as variáveis

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

# Removemos a espécie porque ela não é numerica, entao não é possivel a categorizar com outros valores numericos
# .values é opcional
eixo.boxplot(df.drop('espécie', axis=1).values)

eixo.set_title('Distribuição dos dados', fontsize=15, pad=10)

eixo.set_xticklabels(df.drop('espécie', axis=1).columns)
```

## Adicionando cores

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

# patch_artist preenche as caixas com cores
eixo.boxplot(df.drop('espécie', axis=1), patch_artist=True)

eixo.set_title('Distribuição dos dados', fontsize=15, pad=10)

eixo.set_xticklabels(df.drop('espécie', axis=1).columns)
```

## Alterando cores nas caixas

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

# Definimos as cores
cores = ['pink', 'cyan', 'purple' ,'green']

caixas = eixo.boxplot(df.drop('espécie', axis=1), patch_artist=True)

eixo.set_title('Distribuição dos dados', fontsize=15, pad=10)

eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

# Passando por cada caixa, definimos a cor dela de acordo com a lista
for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)
```

## (Extra) Adicionando marcadores de média

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

cores = ['pink', 'cyan', 'purple' ,'green']

# Ativamos a marcação de média e setamos para ser uma linha
caixas = eixo.boxplot(df.drop('espécie', axis=1), patch_artist=True, showmeans=True, meanline=True)

# Alteramos tipo e cor da linha
for median in caixas['means']:
    median.set_color('red')
    median.set(linestyle='-.')

eixo.set_title('Distribuição dos dados', fontsize=15, pad=10)

eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)
    
```

## Alterando tamanho e tipo de marcadores de outliers

```python
fig = plt.figure(figsize=(8, 5))
eixo = fig.add_axes([0, 0, 1, 1])

cores = ['pink', 'cyan', 'purple' ,'green']

caixas = eixo.boxplot(df.drop('espécie', axis=1), patch_artist=True)

eixo.set_title('Distribuição dos dados', fontsize=15, pad=10)

eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)

for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
```

## Criando Histogramas

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df['comprimento_pétala'])
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

## Aumentando quantidade de barras para aumentar os detalhes do histograma

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

## (Extra) Alterando orientação de histograma

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df['comprimento_pétala'], bins=20, orientation='horizontal')
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_ylabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

## Normalizando o histograma (Para deixar comparação mais fácil)

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df['comprimento_pétala'], bins=20, density=True)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)
```

## (Extra) histograma e boxplot juntos

```python
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, .5])
eixo2 = fig.add_axes([0, .5, 1, .5])

eixo.hist(df['comprimento_pétala'], bins=20, density=True)
eixo.set_title('Distribuição do comprimento das Pétalas', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.grid(True)

eixo2.boxplot(df['comprimento_pétala'],vert=False)
```
